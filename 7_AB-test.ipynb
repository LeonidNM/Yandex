## <a name="Часть1">[Часть 1. Приоритизация гипотез](#Принятие)</a>  

При планировании любого вида деятельности всегда необходимо исходить из существующих ограничений - в том числе и временных. Учесть эти ограничения позволяет выбор приоритетных задач - то есть, задач которые должны быть решены в первую очередь, т.к. являются критически важными. При этом, выбор приоритетных задач не всегда является очевидным - на степень важности влияет совокупность факторов, которые необходимо учесть.

### <a name="1.1">[1.1 Примените фреймворк ICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета](#Принятие)</a>

ICE - субъективный метод оценки важности факторов, где каждому из них присваивается оценка по 10-балльно1 шкале.  
Impact (Влияние) - показывает, насколько идея положительно повлияет на ключевой показатель, который вы пытаются улучшить.  
Confidence (Уверенность в оценке) - показывает, степень уверенности в оценках влияния и легкости реализации.  
Effort (Трудозатраты)​.  

```python
def analized_ice(hypothesis):
    new_hypot = cp.deepcopy(hypothesis)
    new_hypot['ICE'] = new_hypot['Impact'] * new_hypot['Confidence'] / new_hypot['Efforts']
    return new_hypot.round(2).sort_values(by = 'ICE', ascending = False)
```

Как показала практика, работа над проектами занимает много времени, и ведется она с большими перерывами - так что к моменту окончания работы совершенно забывается, что было вначале. В процессе работы в исходные таблицы добавляются новые данные, производные от уже существующих. Эти сущности множатся и однажды, в совершенно неожиданом месте можно получить ошибку. И чтобы ее выявить - необходимо перелопатить весь код от начала проекта. Во избежание такого рода неприятностей код в данном проекте будет, по мере возможности, заключаться в фуекции. Все создаваемые объекты остаются внутри использующих их функций и не влияют на вычисления в остальном проекте.

```python
analized_ice(hypothesis)
```

В результате применения аналитической функции, в таблице hypothesis добавлен новый столбец с информации об оценке  каждой гипотезы по методу ICE. И наивысшую оценку получило предложение "Запустить акцию, дающую скидку на товар в день рождения".

### <a name="1.2">[1.2 Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета](#Принятие)</a>

В дополнение к Impact, Confidence и Effort учитывается также и Reach (Уровень охвата)

```python
def analized_rice(hypothesis):
    new_hypot = analized_ice(hypothesis)
    new_hypot['RICE'] = new_hypot['Reach'] * new_hypot['Impact'] * new_hypot['Confidence'] / new_hypot['Efforts']
    return new_hypot.round(2).sort_values(by = 'RICE', ascending = False)
```

```python
analized_rice(hypothesis)
```

Добавление нового фактора совершенно изменило картину. На первое место вышло предложение "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок".

### <a name="1.3">[1.3 Укажите, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объясните, почему так произошло](#Принятие)</a>

Как было отмечено, при оценке по методу ICE первое место заняло предложение "Запустить акцию, дающую скидку на товар в день рождения". При оценке по методу RICE выяснилось, что у данного предложения будет маленький охват, и оно опустилось в середину списка. Зато, благодаря максимальной оценке охвата, на первое место вышло предложение "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", которое в предыдущий раз было всего лишь на третьем месте. 

Поскольку оценка важности тех или иных факторов в баллах носит субъективный характер (по крайней мере, здесь не раскрыта методика, по которой определяли значимость того или иного фактора), то можно предположить что если добавить еще какой-либо фактор - то рейтинг предложений вновь кардинальным образом поменяется.
Вероятно, следует быть крайне осторожным при выборе методики оценки приоритетности задач.

А/В-тестирование заключается в сравнении показателей контрольной группы А и групппы испытуемых В. Для чистоты эксперимента следует исключить возможность одновременного участия и в контрольной группе и в группе испытателей. 

```python
def analized_orders(orders):
    ord = cp.deepcopy(orders)
    display('Всего в таблице {} строк'.format(ord.shape[0]))
    display('В том числе по группам:')
    ord_a = ord.query('group == "A"')
    display('A - {} строк'.format(ord_a.shape[0]))
    ord_b = ord.query('group == "B"')
    display('B - {} строк'.format(ord_b.shape[0]))
    display('Всего {} уникальных пользователей'.format(ord['visitorId'].nunique()))
    display('В том числе по группе А: {}'.format(ord_a['visitorId'].nunique()))
    display('В том числе по группе B: {}'.format(ord_b['visitorId'].nunique()))
    if ord_a['visitorId'].nunique() + ord_b['visitorId'].nunique() != ord['visitorId'].nunique():
        ord_a_b = ord.groupby('visitorId').agg({'group':'nunique'})
        ord_a_b = ord_a_b.query('group > 1')
        ord_a_b_count = ord_a_b.shape[0]
        display('Разница между числом уникальных пользователей по группам и их общей численностью - {} человек'
                .format(ord_a_b_count))
```

```python
analized_orders(orders)
```

В таблице orders приводится информация о покупках, которые были сделаны участниками тестовых групп, где каждая строка - это одна покупка. Число строк по таблице больше чем число уникальных пользователей. Это указывает на наличие пользователей совершивших более одной покупки. Число участников групп А и В, в совокупности, больше чем число уникальных пользователей на 58 человек. Это указывает на участие этих 58 человек в обеих группах, что является нарушением требования о чистоте эксперимента. 

```python
ord_a_b = orders.groupby('visitorId').agg({'group':'nunique'}).reset_index()
ord_a_b = ord_a_b.query('group > 1')
dict_ord = ord_a_b['visitorId'].tolist()
orders = orders.query('visitorId not in @dict_ord')
analized_orders(orders)
```

Из таблицы orders удалена информация о 58 (примерно 5 процентов от первоначального числа) участниках эксперимента, находящихся сразу в двух группах. Это привело к автоматическому удалению информации о 181 покупке (около 10 процентов от первоначального числа покупок), которые были совершены нарушителями правил.

### <a name="2.1">[2.1 Постройте график кумулятивной выручки по группам. Сделайте выводы и предположения](#Принятие)</a>

Чтобы исключить проблему подсматривания, изучают графики кумулятивных, или накапливаемых данных. Если построить график по кумулятивным данным, в точке первого дня будут значения метрик за этот день, в точке второго дня — набор метрик за два дня, в точке третьего — набор за три. Так отслеживают изменения результатов эксперимента на каждый день тестирования. По центральной предельной теореме значение кумулятивных метрик часто сходится и устанавливается около некоторого среднего.  

```python
def cum_rev(orders):
    ord = cp.deepcopy(orders)
    ord_a = ord.query('group == "A"')
    cum_a = ord_a.groupby(['date'])['revenue'].sum().cumsum().to_frame()
    cum_a = cum_a.reset_index().set_index('date')
    ord_b = ord.query('group == "B"')
    cum_b = ord_b.groupby(['date'])['revenue'].sum().cumsum().to_frame()
    cum_b = cum_b.reset_index().set_index('date')
    ax = plt.gca()
    title = 'Кумулятивная выручка по группам'
    cum_a['revenue'].plot(figsize = (17,9), 
                          grid = True,
                          ax = ax, 
                          color = 'green',
                          title = title)
    cum_b['revenue'].plot(figsize = (17,9),
                          grid = True,
                          ax = ax,
                          color = 'red',
                          title = title)
    plt.legend(['Выручка по группе А', 
                'Выручка по группе В'])
    plt.xlabel('Период времени')
    plt.ylabel('Выручка')
    plt.show()
```

```python
cum_rev(orders)
```

Наблюдения велись в период с перового августа 2019 года по 31 августа 2019 года. И если первой половине этого периода графики кумулятивной выручки по группам были нестабильны (нельзя было однозначно отдать предпочтение какой-то одной группе), то 18 или 19 августа 2019 года произошел скачок выручки по группе В. Далее, примерно 19 августа 2019 года, ежедневная выручка стабилизировалась, графики размеров выручки групп на этом участке следуют практически парралельно друг другу, но благодаря указанному скачку группе В по итогам наблюдений обеспечена победа. У группы А скачков не было, график относительно монотонно возрастает.

### <a name="2.2">[2.2 Постройте график кумулятивного среднего чека по группам. Сделайте выводы и предположения](#Принятие)</a>

Создается составная функция, которая состоит из функции, вычисляющей табличные данные; функции создающей график и сборной функции, вызывающей первые две функции.

```python
def cum_mean_group(orders):
    ord = cp.deepcopy(orders)
    date_group = ord[['date',
                      'group']].sort_values(by = ['date', 
                                                  'group'])
    cum = date_group.apply(lambda x: ord[np.logical_and(x['date'] >= ord['date'], 
                                                                     x['group'] == ord['group'])]\
                           .agg({'date':'max', 
                                 'group':'max', 
                                 'transactionId':'nunique',  
                                 'revenue':'sum'}), 
                           axis = 1)
    cum.columns = ['date', 
                   'group', 
                   'orders', 
                   'revenue']
    cum_a = cum[cum['group'] == 'A']
    cum_b = cum[cum['group'] == 'B']
    return cum_a, cum_b
```

```python
def plot_cum(cum_a, cum_b):
    plt.figure(figsize = (17,9))
    plt.grid(True)
    plt.plot(cum_a['date'], cum_a['revenue'] / cum_a['orders'], 
             color = 'green',
             label = 'A')
    plt.plot(cum_b['date'], cum_b['revenue'] / cum_b['orders'], 
             color = 'red', 
             label = 'B')
    plt.title('Графики кумулятивного среднего чека по группам A/B')
    plt.ylabel('Средний чек')
    plt.xlabel('Дата')
    plt.xticks(rotation = 90)
    plt.legend(['Средний чек по группе А', 
                'Средний чек по группе В'])
    plt.show()
```

```python
def cum_mean(orders):
    cum_a, cum_b = cum_mean_group(orders)
    plot_cum(cum_a, cum_b)
```

```python
def rel_check(orders, visitors):
    ord = cp.deepcopy(orders)
    vis = cp.deepcopy(visitors)
    ord['date'] = ord['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
    vis['date'] = vis['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
    dateGroup = ord[['date','group']].drop_duplicates()
    ordAgg = dateGroup.apply(lambda x: ord[np\
                                           .logical_and(ord['date'] <= x['date'], 
                                                        ord['group'] == x['group'])]\
                             .agg({'date' : 'max', 
                                   'group' : 'max', 
                                   'transactionId' : pd.Series.nunique, 
                                   'visitorId' : pd.Series.nunique, 
                                   'revenue' : 'sum'}), axis = 1)\
    .sort_values(by = ['date',
                       'group'])
    visAgg = dateGroup.apply(lambda x: vis[np\
                                            .logical_and(vis['date'] <= x['date'], 
                                                         vis['group'] == x['group'])]\
                              .agg({'date' : 'max', 
                                    'group' : 'max', 
                                    'visitors' : 'sum'}), axis = 1)\
    .sort_values(by = ['date',
                       'group'])
    cumData = ordAgg.merge(visAgg, 
                           left_on = ['date', 
                                      'group'], 
                           right_on = ['date', 
                                       'group'])
    cumData.columns = ['date', 
                       'group', 
                       'orders', 
                       'buyers', 
                       'revenue', 
                       'visitors']
    cumRevA = cumData[cumData['group'] == 'A'][['date',
                                                'revenue', 
                                                'orders']]
    cumRevB = cumData[cumData['group'] == 'B'][['date',
                                                'revenue', 
                                                'orders']]
    mergCumRev = cumRevA.merge(cumRevB, 
                               left_on = 'date', 
                               right_on = 'date', 
                               how = 'left', 
                               suffixes = ['A', 
                                           'B'])
    plt.figure(figsize = (17,9))
    plt.grid(True)
    plt.plot(mergCumRev['date'], 
             (mergCumRev['revenueB']/mergCumRev['ordersB'])/(mergCumRev['revenueA']/mergCumRev['ordersA'])-1)
    plt.axhline(y = 0, 
                color = 'black', 
                linestyle = '--')
    plt.title('График относительного изменения кумулятивного среднего чека группы B к чеку группы A')
    plt.xlabel('Дата')
    plt.ylabel('Отношение чека группы В к чеку группы А')
    plt.show()
```

### <a name="2.10">[2.10 Посчитайте статистическую значимость различий в конверсии между группами по «сырым» данным. Сделайте выводы и предположения](#Принятие)</a>

В статистике величину (значение) переменной называют статисти́чески зна́чимой, если мала вероятность случайного возникновения этой или ещё более крайних величин.

"Сырые" данные здесь - данные содержащие инфориацию о "аномальных" пользователях (то есть, пользоваелях у которых более 1 покупки, цена заказов у которых превышает 26 785 у.е.).

```python
def stat_convers(df, visitors):
    ord = cp.deepcopy(df)
    vis = cp.deepcopy(visitors)
    ord_a = ord[ord['group'] == 'A'].groupby('visitorId', as_index = False).agg({'transactionId':'nunique'})
    ord_a.columns = ['visitorId', 'orders']
    ord_b = ord[ord['group'] == 'B'].groupby('visitorId', as_index = False).agg({'transactionId':'nunique'})
    ord_b.columns = ['visitorId', 'orders']
    list_ord_a =[]
    for i in range(0,(vis[vis['group'] == 'A']['visitors'].sum() - len(ord_a))):
        list_ord_a.append(0)
    ord_non_purch_users_a = pd.Series(data = list_ord_a, name = 'orders')
    list_ord_b =[]
    for i in range(0,(vis[vis['group'] == 'B']['visitors'].sum() - len(ord_b))):
        list_ord_b.append(0)
    ord_non_purch_users_b = pd.Series(data = list_ord_b, name = 'orders')
    samp_a = pd.concat([ord_a['orders'], ord_non_purch_users_a], axis = 0)
    samp_b = pd.concat([ord_b['orders'], ord_non_purch_users_b], axis = 0)
    display('Нулевая гипотеза - различия в конверсии между группами A и B статистически значимы')
    display('Альтернативная гипотеза - различия в конверсии между группами A и B отсутствуют')
    display('Относительный прирост конверсии группы B по отношению к группе A: {:.3f}'\
            .format(samp_b.mean() / samp_a.mean()-1))
    alpha = 0.05
    result = stats.mannwhitneyu(samp_a, samp_b, alternative = 'two-sided')
    display('p-value: {}'.format(result.pvalue))
    if result.pvalue < alpha:
        display('Нулевая гипотеза подтверждена. Различия в конверсии между группами A и B статистически значимы')
    else:
        display('Нулевая гипотеза отклонена. Различия в конверсии между группами A и B отсутствуют')
```

```python
stat_convers(orders, visitors)
```

Ранее уже было отмечено, что конверсия у группы В выше, чем у группы А.  
Эти данные были получены по таблицам, которые содержали информацию по всем пользователем - и "обычным" и "аномальным".  
Также, было отмечено, что по некоторым позициям полученные данные были похожи на случайные.  
Однако, в рассматриваемом случае, различия в конверсии между группами В и А не являются случайными )то есть могут быть использованы для того чтобы делать обоснованные выводы).

### <a name="2.11">[2.11 Посчитайте статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделайте выводы и предположения](#Принятие)</a>

```python
def stat_check(df):
    ord = cp.deepcopy(df)
    ord[ord['group'] == 'B']['revenue'].mean() / ord[ord['group'] == 'A']['revenue'].mean() - 1
    alpha = 0.05
    display('Нулевая гипотеза - различия в среднем чеке заказа между группами A и B статистически значимы')
    display('Альтернативная гипотеза - различия в среднем чеке заказа между группами A и B отсутствуют')
    result = stats.mannwhitneyu(ord[ord['group'] == 'A']['revenue'], 
                                ord[ord['group'] == 'B']['revenue'], 
                                alternative = 'two-sided')
    display('p-value: {}'.format(result.pvalue))
    if result.pvalue < alpha:
        display('Различия в среднем чеке заказа между группами A и B статистически значимы')
    else:
        display('Различия в среднем чеке заказа между группами A и B отсутствуют')
```
